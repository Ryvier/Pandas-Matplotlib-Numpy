```python
# Sample Trend
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\enzri\OneDrive\Desktop\projects\test.csv")
# Define the 'lamp_id' you want to plot (change to your desired lamp_id)
lamp_id = 99

# Filter data for the selected 'lamp_id'
lamp_data = df[df['lamp_id'] == lamp_id]

# Transpose the data to have dates as rows and measurements as columns
lamp_data = lamp_data.T[1:]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(lamp_data.index, lamp_data.values, marker='o', linestyle='-')
plt.title(f'Trend for lamp_id {lamp_id}')
plt.xlabel('Date')
plt.ylabel('Variable Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Sample Trend Lowest Activity (Uncleaned)
# Set the 'lamp_id' column as the index
df.set_index('lamp_id', inplace=True)

# Calculate the sum of values for each lamp_id
lamp_activity_sum = df.sum(axis=1)

# Find the lamp_id with the lowest activity
lowest_activity_lamp_id = lamp_activity_sum.idxmin()

# Access the data for the lamp_id with the lowest activity
data = df.loc[lowest_activity_lamp_id][1:]

# Create a line plot to represent the trend
plt.figure(figsize=(12, 6))
plt.plot(data.index, data.values, marker='o', linestyle='-')
plt.title('Trend for the Lowest Activity')
plt.xlabel('Date')
plt.ylabel('Activity')
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
# Sample Trend Highest Activity (Uncleaned)
# Find the highest activity for each date
highest_activity = df.iloc[:, 1:].max()

# Create a line plot for the highest activity
plt.figure(figsize=(12, 6))
plt.plot(highest_activity.index, highest_activity.values, marker='o', linestyle='-')
plt.title('Trend for the Highest Activity')
plt.xlabel('Date')
plt.ylabel('Highest Activity')
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```
